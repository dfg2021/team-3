```python
goal4_focus = goal4_focus.rename(columns = {'Education level': 'Educationlevel', 'Reporting Type': 'ReportingType', 'Type of skill': 'TypeOfSkill'}, inplace = False)
goal4_focus.columns
```

```python
goal_441 = goal4_focus[goal4_focus['Indicator'] == '4.4.1']
```

```python
goal4_focus.head()
```

```python
import scipy.stats as stats
from bioinfokit.analys import stat
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
goal_441.columns
```

```python
goal_441['Nature'].unique()
```

```python
# # Ordinary Least Squares (OLS) model
# model_def = 'Value ~ C(TimePeriod) + C(GeoAreaName) + C(Age)'
model_def = 'Value ~ C(TimePeriod) + C(GeoAreaName)  + C(Nature) + C(Sex) + C(TypeOfSkill) + C(Sex):C(TypeOfSkill)'
model = ols(model_def, data=goal4_focus).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
# print(anova_table)

# # ANOVA table using bioinfokit v1.0.3 or later (it uses wrapper script for anova_lm)

res = stat()
res.anova_stat(df=goal4_focus, res_var='Value', anova_model=model_def)
res.anova_summary
```

```python
quantiles = goal_441[["GeoAreaName", "Sex", "Value"]].groupby(["GeoAreaName", "Sex"]).quantile([0.25, 0.75]).reset_index()
quantiles = quantiles.rename(columns = {"level_2": "Quantile"})
```

```python
quantiles
```

```python
quantile_info = pd.pivot(quantiles, index = ["GeoAreaName"], columns = ['Sex', 'Quantile'],  values=['Value']).reset_index()
```

```python
quantile_info
```

```python
quantile_info["both_sex"] = quantile_info.iloc[:, 2] - quantile_info.iloc[:, 1]
quantile_info["Female"] = quantile_info.iloc[:, 4] - quantile_info.iloc[:, 3]
quantile_info["Male"] = quantile_info.iloc[:, 6] - quantile_info.iloc[:, 5]
quantile_info['Max'] = quantile_info.iloc[:, -3:].max(axis=1)
```

```python
quantile_info.sort_values('Max', ascending=False).head()
```

```python
quantile_info.columns = ['GeoAreaName', 'BOTHSEX_25', 'BOTHSEX_75', 'FEMALE_25', 'FEMALE_75', 'MALE_25', 'MALE_75', 'BOTH_SEX_IQR', 'FEMALE_IQR', 'MALE_IQR', 'MAX']
```

```python
quantile_info.head()
```

```python
quantile_proces_data = pd.melt(quantile_info[['GeoAreaName', 'BOTH_SEX_IQR', 'FEMALE_IQR', 'MALE_IQR']], id_vars=['GeoAreaName'], value_vars = ['BOTH_SEX_IQR', 'FEMALE_IQR', 'MALE_IQR'])
```
